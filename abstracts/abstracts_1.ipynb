```python

import json
import time
import codecs
import requests

def get_data_from_url(url, headers={}):
    r = requests.get(url, headers=headers)
    if not r.ok:
        r.raise_for_status()
    
    return r.text
    #decoded = r.json()
    #return decoded

def get_abstracts_from_pubmed(count, nextCursorMark=None):
    url_pattern = 'https://www.ebi.ac.uk/europepmc/webservices/rest/search?query=mutation&resulttype=core&format=json&pageSize={count}&cursorMark={nextCursorMark}'

    if nextCursorMark is None:
        nextCursorMark='*'

    url = url_pattern.format(count=count, nextCursorMark=nextCursorMark)
    
    print ('URL:')
    print (url)
    
    data = get_data_from_url(url)
#    return [{'title': x['title'], 'abstract': x['abstractText']} for x in data['resultList']['result'] 
#            if 'title'in x and 'abstractText'in x]

    return data

def save_abstracts(hitcount):

    data=get_abstracts_from_pubmed(1000)
    
    fetched = 1000
    output_filename = 'abstracts.json'
    output_f = codecs.open(output_filename, 'w', 'utf-8')
    while True:
        parsed = json.loads(data)
        nextCursorMark=parsed['nextCursorMark']
        
        output_f.write(data + '\n')
        output_f.flush()
        print ('Fetched:', fetched)
        
        if fetched > hitcount:
            break
        
        time.sleep(3)
        data = get_abstracts_from_pubmed(1000,nextCursorMark)
        fetched += 1000
        
        
        
    output_f.close()
        

```

```python
save_abstracts(1000000)
```

```python
import re
import os

def walk():

    # traverse root directory, and list directories as dirs and files as files
    counter = 0
    for root, dirs, files in os.walk("."):
        path = root.split(os.sep)
        #print((len(path) - 1) * '---', os.path.basename(root))
        for filename in files:
            counter += 1
            if counter % 100 == 0:
                print ('Counter: ', counter)
            #print(root, dirs, filename)
            to_yield = os.path.join(root, filename)
            if to_yield.endswith('.txt'):
                yield to_yield

def read_filename(filename):
    with open(filename) as f:
        d = f.read()
        
    return d

def parse_paper(paper):
    
    m = re.search('..........:[cg]\...........', paper)
    if m:
        print (m.group(0))
    
    
def parse_all():
    
    for filename in walk():
        #print ('Filename:', filename)
        paper = read_filename(filename)
        parse_paper(paper)
    
```
